## **Descripción del Problema:**
El juego es desarrollado en un tablero de dimensiones variables, generalmente cuadricular. Encontraremos ciertos elementos que se
distribuirán por todo el tablero, éstos serán:
- Jugador.
- Wumpus
- Tesoro.
- Pozos.
- Existencia de Viento.
- Existencia de Hedor. 

```python
import random

# Definición del laberinto
laberinto = {
    1: [2, 5, 8],
    2: [1, 3, 10],
    3: [2, 4, 12],
    4: [3, 5, 14],
    5: [1, 4, 6],
    6: [5, 7, 15],
    7: [6, 8, 17],
    8: [1, 7, 9],
    9: [8, 10, 18],
    10: [2, 9, 11],
    11: [10, 12, 19],
    12: [3, 11, 13],
    13: [12, 14, 20],
    14: [4, 13, 15],
    15: [6, 14, 16],
    16: [15, 17, 20],
    17: [7, 16, 18],
    18: [9, 17, 19],
    19: [11, 18, 20],
    20: [13, 16, 19]
}

# Definición de los pozos
pozos = random.sample(range(1, 21), 3)

# Inicialización de la ubicación del Wumpus y del jugador
wumpus = random.choice(list(laberinto.keys()))
jugador = random.choice(list(laberinto.keys()))
flechas = 3

# Función para mostrar pistas
def mostrar_pistas(habitacion_actual):
    vecinos = laberinto[habitacion_actual]
    olores = [habitacion for habitacion in vecinos if habitacion == wumpus]
    return olores

# Juego principal
while True:
    print("Estás en la habitación", jugador)
    pistas = mostrar_pistas(jugador)
    
    if jugador == wumpus:
        print("¡El Wumpus te ha atrapado! ¡Perdiste!")
        break
    elif wumpus in pistas:
        print("¡Hueles un Wumpus cercano!")
    else:
        print("No sientes ningún olor a Wumpus cercano.")
    
    if jugador in pozos:
        print("¡Cuidado! Has caído en un pozo. ¡Perdiste!")
        break
    
    opciones_validas = laberinto[jugador]
    print("Opciones de movimiento: ", opciones_validas)
    print("Flechas restantes:", flechas)
    
    accion = input("¿Qué quieres hacer? (m para moverte, f para disparar una flecha, q para salir): ")
    
    if accion.lower() == 'q':
        print("Gracias por jugar. ¡Hasta luego!")
        break
    elif accion.lower() == 'm':
        movimiento = input("¿A qué habitación te gustaría ir? (ingresa el número de la habitación): ")
        if int(movimiento) in opciones_validas:
            jugador = int(movimiento)
        else:
            print("Movimiento inválido. Inténtalo de nuevo.")
    elif accion.lower() == 'f':
        if flechas > 0:
            objetivo = int(input("¿A qué habitación quieres disparar la flecha? "))
            flechas -= 1
            if objetivo == wumpus:
                print("¡Felicidades! Has eliminado al Wumpus. ¡Ganaste!")
                break
            else:
                print("La flecha no alcanzó al Wumpus. El Wumpus se ha movido a otra habitación.")
                wumpus = random.choice(list(laberinto.keys()))
        else:
            print("No te quedan flechas. Encuentra más flechas para seguir disparando.")
    else:
        print("Acción inválida. Inténtalo de nuevo.")
```
